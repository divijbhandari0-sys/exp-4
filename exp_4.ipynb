```python
'''PROBLEM STATEMENT 2: Three Students choose elective subjects.
Find the common subjects chosen by all students using set operation
concept used: Intersection(&)'''
```

```python
# Elective subjects chosen by three stud
stud1_subjects = {"Physics", "Chemistry", "Math", "Computer Science"}
stud2_subjects = {"Math", "Biology", "Chemistry", "English"}
stud3_subjects = {"Computer Science", "Math", "Physics", "History"}

# Find common subjects using set intersection
common_subjects = stud1_subjects & stud2_subjects & stud3_subjects

print("Student 1's subjects:", stud1_subjects)
print("Student 2's subjects:", stud2_subjects)
print("Student 3's subjects:", stud3_subjects)
print("\nCommon subjects chosen by all students:", common_subjects)
```

```python
'''PROBLEM STATEMENT 3: One set contains students in the Cricket club another
in the football club.
find students who are in: 1. Both Clubs 2. In only one club
Concept used: Intersection, Symmetric Difference'''
```
